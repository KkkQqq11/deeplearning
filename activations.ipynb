```python
import torch
```

# Sigmoid

$$S(x) = \frac{1}{1 + e^{-x}}$$

优点：
1、输出稳定在（0,1）
2、连续函数、便于求导
缺点：
1、梯度消失
2、不是零中心的（non-zero centered）：这意味着如果输入的数据总是正的（例如，Sigmoid的输出），那么关于权重的梯度在反向传播时也会总是正的或总是负的，这可能会导致梯度更新的效率降低。
3、求指数计算量大

```python
# 创建一个Tensor
x = torch.range(-5., 5., 0.1)
# 应用sigmoid函数
y = torch.sigmoid(x)

# 打印结果
print(y)

# 使用matplotlib可视化结果
import matplotlib.pyplot as plt

plt.plot(x.numpy(), y.numpy())
plt.title("Sigmoid Function")
plt.xlabel("x")
plt.ylabel("Sigmoid(x)")
plt.grid(True)
plt.show()
```

# Tanh
优点:
零中心化（zero-centered），这使得数据在经过激活函数之后的分布有正有负，有助于提高梯度下降的效率。
缺点:
与Sigmoid函数一样，Tanh函数也有梯度消失的问题，尤其是在输入值很大或很小的时候。

```python
# 创建一个Tensor
x = torch.range(-5., 5., 0.1)
# 应用sigmoid函数
y = torch.tanh(x)

# 打印结果
print(y)

# 使用matplotlib可视化结果
import matplotlib.pyplot as plt

plt.plot(x.numpy(), y.numpy())
plt.title("Tanh Function")
plt.xlabel("x")
plt.ylabel("Tanh(x)")
plt.grid(True)
plt.show()
```

# Relu
优点:
1、计算简单，加速了神经网络的训练。
2、解决了梯度消失问题（在正区间）。
3、在实践中被证明非常有效，是目前最流行的激活函数之一。
缺点:
1、死亡ReLU问题：如果一个神经元的输入总是负的，那么ReLU激活后的输出总是0，这个神经元就可能再也不会被激活，因此在训练过程中“死掉”。
2、不是零中心化的。

```python
# 创建一个Tensor
x = torch.range(-5., 5., 0.1)
# 应用sigmoid函数
y = torch.relu(x)

# 打印结果
print(y)

# 使用matplotlib可视化结果
import matplotlib.pyplot as plt

plt.plot(x.numpy(), y.numpy())
plt.title("Relu Function")
plt.xlabel("x")
plt.ylabel("Relu(x)")
plt.grid(True)
plt.show()
```

# Leaky Relu
优点:
尝试解决ReLU的死亡问题，通过允许负数输入有一个非零的小梯度。
缺点:
需要调整额外的参数（PReLU）或选择合适的“泄漏”参数（Leaky ReLU）。

```python
activation = torch.nn.LeakyReLU()
# 创建一个Tensor
x = torch.range(-5., 5., 0.1)
# 应用sigmoid函数
y = activation(x)

# 打印结果
print(y)

# 使用matplotlib可视化结果
import matplotlib.pyplot as plt

plt.plot(x.numpy(), y.numpy())
plt.title("Leaky Relu Function")
plt.xlabel("x")
plt.ylabel("Leaky Relu(x)")
plt.grid(True)
plt.show()
```

# Swish函数

优点:
被证明在多种任务中比ReLU表现得更好。
具有自我正则化的性质，有助于网络的训练。
缺点:
计算上比ReLU复杂。

```python
activation = torch.nn.SiLU()
# 创建一个Tensor
x = torch.range(-5., 5., 0.1)
# 应用sigmoid函数
y = activation(x)

# 打印结果
print(y)

# 使用matplotlib可视化结果
import matplotlib.pyplot as plt

plt.plot(x.numpy(), y.numpy())
plt.title("Swish Function")
plt.xlabel("x")
plt.ylabel("Swish(x)")
plt.grid(True)
plt.show()
```
