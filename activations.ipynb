# Sigmoid

$$S(x) = \frac{1}{1 + e^{-x}}$$

$$
\begin{align*}
& 优点：\\
& 1、输出稳定在（0,1）\\
& 2、连续函数、便于求导\\
&  缺点：\\
& 1、梯度消失\\
& 2、关于原点不对称\\
& 3、求指数计算量大\\
\end{align*}
$$

```python
import torch

# 创建一个Tensor
x = torch.range(-5., 5., 0.1)
# 应用sigmoid函数
y = torch.sigmoid(x)

# 打印结果
print(y)

# 使用matplotlib可视化结果
import matplotlib.pyplot as plt

plt.plot(x.numpy(), y.numpy())
plt.title("Sigmoid Function")
plt.xlabel("x")
plt.ylabel("Sigmoid(x)")
plt.grid(True)
plt.show()
```

```python
# 创建一个Tensor
x = torch.range(-5., 5., 0.1)
# 应用sigmoid函数
y = torch.relu(x)

# 打印结果
print(y)

# 使用matplotlib可视化结果
import matplotlib.pyplot as plt

plt.plot(x.numpy(), y.numpy())
plt.title("Relu Function")
plt.xlabel("x")
plt.ylabel("Relu(x)")
plt.grid(True)
plt.show()
```
